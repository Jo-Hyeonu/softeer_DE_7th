```python
#Orders table

cur.execute("""
    CREATE TABLE IF NOT EXISTS Orders
        (OrderID INTEGER PRIMARY KEY,
        CustomerID INTEGER,
            EmployeeID INTEGER,
            OrderDate TEXT,
            ShipperID INTEGER
        )
            """)
orders_data = [
        (10248, 92, 5, '1996-07-04', 3),
        (10249, 81, 6, '1996-07-05', 1),
        (10250, 93, 4, '1996-07-08', 2),
        (10251, 84, 3, '1996-07-09', 1),
        (10252, 76, 4, '1996-07-10', 2)
    ]
cur.executemany("""
        INSERT OR IGNORE INTO Orders 
        (OrderID, CustomerID, EmployeeID, OrderDate, ShipperID) 
        VALUES (?, ?, ?, ?, ?)
    """, orders_data)
```

```python
#inner join
cur.execute("SELECT o.OrderID, c.CustomerName, o.OrderDate FROM Orders as o INNER JOIN Customers as c ON o.CustomerID = c.CustomerID")

data = cur.fetchall()
for row in data:
    print(row)
```

```python
#left join
cur.execute("SELECT c.CustomerName, o.OrderID FROM Customers as c left join Orders as o on c.CustomerID = o.CustomerID ORDER BY c.CustomerName")
data = cur.fetchall()
for row in data:
    print(row)
```

```python
#right Join
cur.execute("SELECT c.CustomerName, o.OrderID FROM Customers as c right join Orders as o on c.CustomerID = o.CustomerID ORDER BY c.CustomerName")
data = cur.fetchall()
for row in data:
    print(row)
```

```python
#full outer JOin
#FULL OUTER JOIN , FULL JOIN 두 개 같음.
cur.execute("SELECT c.CustomerName, o.OrderID FROM Customers as c full outer join Orders as o on c.CustomerID = o.CustomerID ORDER BY c.CustomerName")
data = cur.fetchall()
for row in data:
    print(row)
```

```python
#self join
cur.execute("select a.CustomerName , b.CustomerName as CustomerName2, a.City from Customers A, Customers B where a.CustomerID <> b.CustomerID AND a.City == b.City Order by a.City")
data = cur.fetchall()
for row in data:
    print(row)
```

```python
#union

cur.execute("""
    select CustomerID FROM Customers
    union
    select CustomerID FROM Orders
""")
data = cur.fetchall()
for row in data:
    print(row)
```

```python
#union all
cur.execute("""
    select CustomerID FROM Customers
    union all
    select CustomerID FROM Orders
    order by CustomerID
""")
data = cur.fetchall()
for row in data:
    print(row)
```

```python
#group by
cur.execute("""
    select count(CustomerID), Country
    from Customers
    group by Country
    order by Count(CustomerID) DESC  

""")
data = cur.fetchall()
for row in data:
    print(row)
```

```python
#having
cur.execute("""
    select count(CustomerID), Country
    from Customers
    group by Country
    having Count(CustomerID) > 1
    order by Count(CustomerID) DESC  

""")
data = cur.fetchall()
for row in data:
    print(row)
```
